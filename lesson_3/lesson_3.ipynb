## Тема 3 “Последовательность”

Даны 4 последовательности. Необходимо:
* исследовать их на монотонность; 
* исследовать на ограниченность;
* найти пятый по счету член.

```python
from IPython.display import Image
Image('task.png')
```

```python
a, b, c, d = [],[],[],[]
for n in range(1, 6):
    a.append(2**n - n)
    c.append(-1**n + pow(2*n, 0.5))
    d.append(pow(-1, 2*n) + 1/n/n)
for n in range(2, 7):
    b.append(1/(1-n))
```

```python
for i in (a,b,c,d):
    print(i)
```

* последовательность_1 - монотонная, неограниченная
* последовательность_2 - монотонная, ограниченная
* последовательность_3 - монотонная, неограниченная
* последовательность_4 - монотонная, ограниченная

```python
for ind,el in enumerate([a,b,c,d]):
    print(f'sequence_{ind + 1}, n=5: {el[4]}')
```
