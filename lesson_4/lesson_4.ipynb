```python
import matplotlib.pyplot as plt
import numpy as np

from IPython.display import Image
```

__Задание 1__

```python
def f(x):
    return [np.cos(el) if el <= 0 else np.sin(el) for el in x]

x = np.linspace(-2*np.pi, 2*np.pi, 100)
x = np.append(x, 0)
x.sort()
```

```python
plt.plot([el for el in x if el > 0], f([el for el in x if el > 0]), 'b')
plt.plot([el for el in x if el <= 0], f([el for el in x if el <= 0]), 'r')
ax = plt.gca()
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.show()
```

```python
def ctg(x):
    return [np.nan if el % np.pi ==0 else np.cos(el)/np.sin(el) for el in x]

x = np.linspace(-np.pi, np.pi, 1000)
x = np.append(x, 0)
x.sort()
```

```python
plt.plot(x, ctg(x))
ax = plt.gca()
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.ylim(-5,5)
plt.show()
```

```python
def f(x):
    return x**3 - x**2

x = np.arange(-1, 2, 0.01)
```

```python
plt.plot(x, [f(x_i) for x_i in x])
ax = plt.gca()
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.show()
```
