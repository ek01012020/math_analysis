## Тема 6 “Понятие о производной”

```python
from IPython.display import Image
```

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sym

%matplotlib inline
x, x_k = sym.symbols('x, x_k')
```

```python
def func_derivative(y):
    yd = sym.diff(y, x)
    
    list_x = np.arange(0.1,10,0.1)
    list_y = [y.subs(x, i) for i in list_x]
    list_yd = [yd.subs(x, i) for i in list_x]
    
    fig = plt.figure(figsize=(20, 12))
    
    ax1 = fig.add_subplot(211)
    ax1.plot(list_x, list_y, 'b', label='function')
    ax1.plot(list_x, list_yd, 'r', label='derivative')
    ax1.legend()
    
    ax2 = fig.add_subplot(223)
    ax2.plot(list_x, list_y, 'b', label='function')
    ax2.legend()
    
    ax3 = fig.add_subplot(224)
    ax3.plot(list_x, list_yd, 'r', label='derivative')
    ax3.legend()
    plt.show()
```

```python
y = (x**3-x**2-x-1)/(1+2*x+3*x**2-4*x**3)
func_derivative(y)
```
