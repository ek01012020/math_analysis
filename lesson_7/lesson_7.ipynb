```python
y_ = sym.solve(2*(x+y)-144, y)
y_
```

```python
s = s.subs(y, y_[0])
s
```

```python
def extremum(func):
    diff_f = sym.diff(func, x)
    try:
        if not sym.roots(diff_f, x):
            return print('–ù–µ—Ç –∫—Ä–∏—Ç–∏—á–µ—Å–∫–∏—Ö —Ç–æ—á–µ–∫ –¥–ª—è –ø–æ–∏—Å–∫–∞ —ç–∫—Å—Ç—Ä–µ–º—É–º–∞')
        roots = sym.roots(diff_f, x)
        r_min = min(sym.roots(diff_f, x).keys())
        r_max = max(sym.roots(diff_f, x).keys())
        sym.plot(func, (x, r_min-1, r_max+1), title='Function')
        sym.plot(diff_f, (x, r_min-1, r_max+1), title='Derivative')
        print(f"–ö–æ—Ä–Ω–∏ —É—Ä–∞–≤–Ω–µ–Ω–∏—è: {list(roots.keys())}")
    except Exception:
        sym.plot(func, title='Function')
        print('–ü—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –Ω–µ –∏–º–µ–µ—Ç –Ω—É–ª–µ–π')
```

```python
extremum(s)
```
